```python
import pandas as pd
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
```

```python
data = pd.read_csv('hitdata_working.txt', delimiter='\t', index_col=1)
display(data)
```

```python
record_dict = SeqIO.index("158_NS1.fasta", "fasta")
print(record_dict['AAB23699.1'].description)
print(record_dict['AAB23699.1'].seq)
```

```python
new_dict = []
for i in data.index:
    name = str(i.split()[0])
    start = data['From'][i]
    end = data['To'][i]
    new_string = record_dict[name].seq[start:end]
    new_dict.append(SeqRecord(Seq(new_string), id=name,name=name, description=record_dict[name].description))
print('done :)')
```

```python
SeqIO.write(new_dict, "fragments.fasta", "fasta")
```
